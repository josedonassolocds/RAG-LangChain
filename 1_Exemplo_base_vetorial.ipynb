```python
import chromadb
from pprintpp import pprint
```

```python
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.get_or_create_collection(name="semana_academica")
```

```python
collection.upsert(
    documents=[
        "A semana academica acontece no Brasil.",
        "A semana academica acontece na França.",
        "A semana academica acontece no Japão.",
        "A semana academica acontece na Alemanha.",
        "A semana academica acontece nos Estados Unidos.",
        "A semana academica acontece no Canadá.",
        "A semana academica acontece na Austrália.",
        "A semana academica acontece na Itália.",
        "A semana academica acontece na Argentina.",
        "A semana academica acontece na Espanha.",
        "A semana academica acontece na Rússia."
    ],
    ids=["pais1", "pais2", "pais3", "pais4", "pais5", "pais6", "pais7", "pais8", "pais9", "pais10", "pais11"]
)
```

```python
resultado = collection.query(
    query_texts=["A semana academica vai acontecer em Roma?"],
    n_results=3,
    # where_document = {'$contains': 'Russia'}
)

pprint(resultado)
```
